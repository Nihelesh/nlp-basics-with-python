```python
def chunk_text(text, chunk_size):
 """
 Divides text into chunks of a predefined size.
 Parameters:
 text (str): The input text to be chunked.
 chunk_size (int): The size of each chunk in characters.
 Returns:
 list: A list of text chunks.
 """
 return [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
# Sample text
text = ("Fixed-size chunking involves dividing the text into chunks of a predefined size, " "typically based on the number of characters or tokens. This method is simple and " "easy to implement, but it may not always capture meaningful units of text." "NLP is simple and easy")
# Chunk size in characters
chunk_size = 50
# Chunk the text
chunks = chunk_text(text, chunk_size)
for i, chunk in enumerate(chunks):
 print(f"Chunk {i+1}:\n{chunk}\n")
```
