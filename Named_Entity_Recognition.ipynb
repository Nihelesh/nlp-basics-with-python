```python
import pandas as pd
import spacy
import requests
from bs4 import BeautifulSoup
nlp = spacy.load("en_core_web_sm")
pd.set_option("display.max_rows", 200)
content = "The Indian Space Research Organisation (ISRO) successfully launched a new weather satellite to improve cyclone forecasting and climate monitoring across the Indian subcontinent."
doc = nlp(content)
for ent in doc.ents:
 print(ent.text, ent.start_char, ent.end_char, ent.label_)
entities = [(ent.text, ent.label_, ent.lemma_) for ent in doc.ents]
df = pd.DataFrame(entities, columns=['text', 'type', 'lemma'])
print(df)
```
